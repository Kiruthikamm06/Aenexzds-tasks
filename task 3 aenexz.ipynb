```python

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Kiruthikamm06/Aenexzds-tasks/main/Iris_data_sample.csv")
print(df.head())  # Display First 5 Rows
print("------------------------------------------------")
print(df.shape) #shape
print("------------------------------------------------")
df.info()
print("------------------------------------------------")
print(df.isnull().sum()) #missing values
print("------------------------------------------------")
print(df.duplicated().sum())



df["SepalLengthCm"] = df["SepalLengthCm"].replace("??", 5.8)     #random values to fill
df["PetalLengthCm"] = df["PetalLengthCm"].replace("###", 3.7)
df["SepalWidthCm"].fillna(df["SepalWidthCm"].mean(), inplace=True)

df["Species"].fillna("Iris-setosa", inplace=True)
print("------------------------------------------------")

df = df.drop("Unnamed: 0", axis=1)

print(df.isnull().sum())
print("------------------------------------------------")
df.to_csv("cleaned_iris.csv", index=False)

print("Data cleaning completed")
```
